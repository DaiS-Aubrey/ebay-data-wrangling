```python
# Convert Price to numeric data 
import re

def process_price(price):
    # Check if the price is a range
    if "to" in price:
        # Extract the two prices using regular expressions and convert them to floats
        prices = re.findall(r'\d+\.?\d*', price)
        if len(prices) == 2:
            price_min = float(prices[0])
            price_max = float(prices[1])
            # Return the average of the two prices
            return (price_min + price_max) / 2
    else:
        # If it's a single price, just remove any non-numeric characters and convert to float
        price = re.sub(r'[^\d.]', '', price)
        if price:
            return float(price)
    return None  # Return None if the price is not in a valid format

# Apply the function to the 'Price' column in df_price
df_price['Price'] = df_price['Price'].apply(process_price)

```

# Extract and Clean RAM Size Data

```python


import numpy as np



# Define a function to extract numerical values from RAM size strings
def extract_ram_size(ram_string):
    if pd.isnull(ram_string):  # Handle missing values
        return np.nan
    # Use regular expression to extract the numeric part
    match = re.search(r'(\d+)', ram_string)
    return float(match.group(1)) if match else np.nan

# Apply the function to extract numerical values from the RAM Size column
df['Ram Size'] = df['Ram Size'].apply(extract_ram_size)

# Fill missing values with the mode (most common value)
most_common_ram = df['Ram Size'].mode()[0]  # Get the mode
df['Ram Size'].fillna(most_common_ram, inplace=True)

```

## Extract and Clean Screen Size Data

```python

# Define a function to extract numerical values from screen size strings
def extract_screen_size(screen_string):
    if pd.isnull(screen_string):  # Handle missing values
        return np.nan
    # Use regular expression to extract the numeric part
    match = re.search(r'(\d+\.?\d*)', screen_string)  # Match integers and decimals
    return float(match.group(1)) if match else np.nan

# Apply the function to extract numerical values from the Screen Size column
df['Screen Size'] = df['Screen Size'].apply(extract_screen_size)

# Fill missing values with the mode (most common value)
most_common_screen_size = df['Screen Size'].mode()[0]  # Get the mode
df['Screen Size'].fillna(most_common_screen_size, inplace=True)

```

```python
# Define a function to extract numerical values from SSD capacity strings
def extract_ssd_capacity(ssd_string):
    if pd.isnull(ssd_string):  # Handle missing values
        return np.nan
    # Use regular expression to extract the numeric part
    match = re.search(r'(\d+\.?\d*)', ssd_string)  # Match integers and decimals
    if match:
        return float(match.group(1))  # Convert to float
    return np.nan  # Return NaN if no match found

# Apply the function to extract numerical values from the SSD Capacity column
df['SSD Capacity'] = df['SSD Capacity'].apply(extract_ssd_capacity)


# Fill missing values with the mode (most common value)
most_common_ssd = df['SSD Capacity'].mode()[0]  # Get the mode
df['SSD Capacity'].fillna(most_common_ssd, inplace=True)


print(df['SSD Capacity'].info())
print(df['SSD Capacity'].head())
```

```python
def convert_tb_to_gb(ssd_value):
    if ssd_value == 1.0:  # If the value is 1 (indicating it's actually 1TB)
        return 1024.0     # Convert 1TB to 1024GB
    if ssd_value == 2.0:
        return 2048.0
    return ssd_value      # Otherwise, return the original value (assumed to be in GB)

# Apply the conversion function to the SSD Capacity column
df['SSD Capacity'] = df['SSD Capacity'].apply(convert_tb_to_gb)
```

## Extract and Clean Hard Drive Capacity Data

```python
# Function to extract numerical values from Hard Drive Capacity strings
def extract_hard_drive_capacity(hd_string):
    if pd.isnull(hd_string):  # Handle missing values
        return np.nan
    # Use regular expression to extract the numeric part
    match = re.search(r'(\d+\.?\d*)', hd_string)  # Match integers and decimals
    if match:
        return float(match.group(1))  # Convert to float
    return np.nan  # Return NaN if no match found

# Apply the function to extract numerical values from the Hard Drive Capacity column
df['Hard Drive Capacity'] = df['Hard Drive Capacity'].apply(extract_hard_drive_capacity)

# Function to convert 1TB values into GB for Hard Drive Capacity
def convert_hd_to_gb(hd_value):
    if hd_value == 1.0:  # If the value is 1 (indicating it's actually 1TB)
        return 1024.0     # Convert 1TB to 1024GB
    if hd_value == 2.0:
        return 2048.0
    return hd_value      # Otherwise, return the original value (assumed to be in GB)

# Apply the conversion function to the Hard Drive Capacity column
df['Hard Drive Capacity'] = df['Hard Drive Capacity'].apply(convert_hd_to_gb)

# Fill missing values with the mode (most common value)
most_common_hd = df['Hard Drive Capacity'].mode()[0]  # Get the mode
df['Hard Drive Capacity'].fillna(most_common_hd, inplace=True)

# Check the results
print(df['Hard Drive Capacity'].value_counts())  # Count of unique Hard Drive capacities
print(df['Hard Drive Capacity'].head())          # Check the first few rows
```

```python
df.head()
```

```python

# Function to extract processor speed from strings
def extract_processor_speed(speed_string):
    if pd.isnull(speed_string):  # Handle missing values
        return np.nan
    # Use regular expression to find the maximum GHz value
    match = re.search(r'(\d+\.?\d*)\s*GHz', speed_string)
    if match:
        return float(match.group(1))  # Convert to float
    return np.nan  # Return NaN if no match found

# Apply the function to extract processor speed
df['Processor Speed'] = df['Processor Speed'].apply(extract_processor_speed)

# Fill missing values with the mode (most common value)
most_common_speed = df['Processor Speed'].mode()[0]  # Get the mode
df['Processor Speed'].fillna(most_common_speed, inplace=True)

# Check the results
print(df['Processor Speed'].value_counts())  # Count of unique processor speeds
print(df['Processor Speed'].head())          # Check the first few rows
```

## Visualization of the Distribution of Numeric Columns: Price, Ram Size, SSD Capacity, and Processor Speed

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Set the style
sns.set(style='whitegrid')

# Function to plot distributions of numeric columns
def plot_numeric_distributions(df, columns):
    plt.figure(figsize=(15, 10))
    
    for i, column in enumerate(columns):
        plt.subplot(2, 2, i + 1)
        
        # Plot the histogram
        sns.histplot(df[column], kde=True, bins=30)
        plt.title(f'Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
    
    plt.tight_layout()
    plt.show()



numeric_columns = ['Price', 'Ram Size (GB)', 'SSD Capacity (GB)', 'Processor Speed (GHz)']

plot_numeric_distributions(df, numeric_columns)

```

## Delete the outliers of Price

```python
price_description = df['Price'].describe()
print("Price Description:\n", price_description)

# Identify outliers 
mean_price = price_description['mean']
std_price = price_description['std']
outlier_threshold_upper = mean_price + 3 * std_price
outlier_threshold_lower = mean_price - 3 * std_price

outliers = df[(df['Price'] > outlier_threshold_upper) | (df['Price'] < outlier_threshold_lower)]
print("Identified Outliers:\n", outliers)


# Remove outliers
df_cleaned = df[~((df['Price'] > outlier_threshold_upper) | (df['Price'] < outlier_threshold_lower))]
```

```python
# 定义标准化函数
def clean_condition(condition):
    if pd.isnull(condition):
        return 'Unknown'  # 处理空值
    condition = condition.lower().strip()   # 转换为小写，方便匹配
    if 'brand-new' in condition:
        return 'New'
    elif 'refurbished' in condition:
        return 'Refurbished'
    elif 'used' in condition:
        return 'Used'
    elif 'parts' in condition:
        return 'For parts or not working'
    else:
        return 'Other'

# 应用函数到 Condition 列
df['Condition'] = df['Condition'].apply(clean_condition)
```
